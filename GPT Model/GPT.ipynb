```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

def generate_health_advice(prompt, model, tokenizer, max_length=150):
    """
    Generates health advice and appointment query handling response.
    """
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, do_sample=True)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Load the fine-tuned GPT model and tokenizer
fine_tuned_model_path = "path_to_fine_tuned_model"  # Replace with your model path
tokenizer = GPT2Tokenizer.from_pretrained(fine_tuned_model_path)
model = GPT2LMHeadModel.from_pretrained(fine_tuned_model_path)

# Example disease prediction and patient context
predicted_disease = "Diabetes"
patient_context = {
    "symptoms": ["fatigue", "frequent urination", "thirst"],
    "age": 45,
    "gender": "male"
}

# Generate prompt for the GPT model
prompt = (
    f"The patient is a {patient_context['age']}-year-old {patient_context['gender']} "
    f"with symptoms {', '.join(patient_context['symptoms'])}. "
    f"The predicted condition is {predicted_disease}. Provide health advice and "
    "suggest a doctor appointment."
)

# Generate health advice and appointment query handling
response = generate_health_advice(prompt, model, tokenizer)

# Output the response
print("Health Advice and Appointment Query Handling:")
print(response)
```
